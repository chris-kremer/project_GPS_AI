```python
def parse_and_format_date(date_str):
    # Convert the input to a string, in case it's not
    date_str = str(date_str)

    # Remove the day name and the period (e.g., 'Wednesday. ') using regex
    date_str = re.sub(r'^.*?\.\s*', '', date_str)
    # Replace ' at ' with a space
    date_str = date_str.replace(' at ', ' ')

    # Convert the string to a datetime object
    date_obj = pd.to_datetime(date_str, errors='coerce')

    # Check if date_obj is NaT before formatting
    if pd.isna(date_obj):
        return None
    else:
        return date_obj.strftime('%Y-%m-%d %H:%M:%S')

def schedule_tasks(df_events, df_tasks):
    now = datetime.now(timezone.utc)
    now = now + timedelta(hours=1)
    today = now.date()
    future_events = df_events[(df_events['start'].dt.date == today) & (df_events['start'] > now)]

    scheduled_tasks = []
    occupied_slots = []

    for _, event in future_events.iterrows():
        occupied_slots.append((event['start'], event['stop']))

    df_tasks = df_tasks.sort_values(by='Adj. Priority', ascending=False)

    for _, task in df_tasks.iterrows():
        start_time = now
        end_time = start_time + timedelta(minutes=task['Estimate'])

        for occupied_start, occupied_stop in occupied_slots:
            if start_time.tzinfo is None:
                occupied_stop_naive = make_naive(occupied_stop)
            else:
                occupied_stop_naive = occupied_stop

            if start_time < occupied_stop_naive + timedelta(minutes=10):
                start_time = occupied_stop_naive + timedelta(minutes=10)
                end_time = start_time + timedelta(minutes=task['Estimate'])

        scheduled_task = task.to_dict()
        scheduled_task['start'] = start_time
        scheduled_task['stop'] = end_time
        scheduled_task['color'] = 'yellow'
        scheduled_task['type'] = 'auto_planned'
        scheduled_task['duration'] = scheduled_task['Estimate']

        scheduled_tasks.append(scheduled_task)

        now = end_time + timedelta(minutes=10)
        occupied_slots.append((start_time, end_time))

    return pd.DataFrame(scheduled_tasks)
def convert_to_datetime_with_default_date(time_val, default_date, timezone='Europe/Berlin'):
    tz = pytz.timezone(timezone)
    if pd.isnull(time_val):
        return None

    # Check if the value is a single datetime object, not a series or array
    if isinstance(time_val, pd.Timestamp) or isinstance(time_val, datetime):
        if time_val.tzinfo is None:
            return tz.localize(time_val)
        else:
            return time_val.tz_convert(timezone)
    elif isinstance(time_val, time):
        return tz.localize(datetime.combine(default_date, time_val))
    else:
        # Handle unexpected types
        print(f"Unexpected type for time_val: {type(time_val)}")
        return None
def plot_activities_of_a_day(df, date):
    # Ensure 'start' is in datetime format
    df['start'] = pd.to_datetime(df['start'])
    
    # Filter data for the specified date
    daily_data = df[df['start'].dt.date == pd.to_datetime(date, format='%d.%m.%Y').date()]
    
    if daily_data.empty:
        print(f"No data for {date}.")
        return
    
    # Sort by start time and calculate duration in hours
    daily_data = daily_data.sort_values(by='start')
    daily_data['duration'] = daily_data['duration'] / 60  # Assuming 'duration' is initially in minutes
    
    # Prepare priority levels and plotting space
    priority_levels = np.sort(daily_data['Priority'].unique())
    fig, ax = plt.subplots(figsize=(12, 6))
    
    texts = []
    bar_patches = []
    
    # Loop through each priority level and plot activities
    for i, priority_level in enumerate(priority_levels):
        df_priority = daily_data[daily_data['Priority'] == priority_level]
        for j in range(len(df_priority)):
            start_hour = df_priority.iloc[j]['start'].hour + df_priority.iloc[j]['start'].minute / 60
            duration = df_priority.iloc[j]['duration']
            
            if df_priority.iloc[j]['color'] == 'transparent':
                facecolor = 'none'
                edgecolor = 'grey'
            else:
                facecolor = df_priority.iloc[j]['color']
                edgecolor = 'black'
            
            bar = ax.broken_barh([(start_hour, duration)], (i - 0.4, 0.8), facecolors=facecolor, edgecolor=edgecolor, linewidth=1)
            bar_patches.append(bar)
            
            if df_priority.iloc[j]['done'] != 52 and df_priority.iloc[j]['Category'] != 'sleep':
                x = start_hour + duration / 2
                y = i
                activity_text = df_priority.iloc[j]['Activity']
                text = ax.text(x, y, activity_text, ha='center', va='center', fontsize=8, color='black', wrap=True, bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.2'))
                texts.append(text)
    
    # Adjust text to prevent overlap
    adjust_text(texts, add_objects=bar_patches, only_move={'texts': 'y'}, force_points=0, force_text=0.5, force_objects=0.5, autoalign=None, expand_text=(1, 1.2), lim=1000)
    
    # Set y-ticks and labels for priority levels
    ax.set_yticks(range(len(priority_levels)))
    ax.set_yticklabels(priority_levels)
    ax.set_xlim(0, 24)  # Assuming 24-hour format for a day

    # Add vertical line for the current time
    # Add vertical line for the current time, adjusted for time zone
    current_time = datetime.now().astimezone().hour + datetime.now().minute / 60
    ax.axvline(x=current_time, color='grey', linestyle='--', linewidth=1, label='Current Time')
    
    # Add legend
    ax.legend()

    plt.tight_layout()
    plt.show()

    
def parse_and_format_date(date_str):
    # Convert the input to a string, in case it's not
    date_str = str(date_str)

    # Remove the day name and the period (e.g., 'Wednesday. ') using regex
    date_str = re.sub(r'^.*?\.\s*', '', date_str)
    # Replace ' at ' with a space
    date_str = date_str.replace(' at ', ' ')

    # Convert the string to a datetime object
    date_obj = pd.to_datetime(date_str, errors='coerce')

    # Check if date_obj is NaT before formatting
    if pd.isna(date_obj):
        return None
    else:
        return date_obj.strftime('%Y-%m-%d %H:%M:%S')
    
```
